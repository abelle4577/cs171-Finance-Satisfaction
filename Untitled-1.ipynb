```python
# import Consumer_Complaints.csv
# import pandas
# create a new csv file that only contains data for dates in 2016 from Date Received column

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('Consumer_Complaints.csv')
df.head()
df.count()
```

```python
# only include data from 2016
df_2016 = df[df['Date received'].str.contains('2016')]
df_2016.head()
```

```python
# count the number of entries in df_2016
df_2016.count()
```

```python
# select and delete 100,000 random rows from df_2016
df_2016 = df_2016.drop(df_2016.sample(100000).index)
df_2016.count()


```

```python
# convert df_2016 to csv
df_2016.to_csv('Consumer_Complaints_2016.csv', index=False)
```
